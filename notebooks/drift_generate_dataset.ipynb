```python
import pandas as pd
import numpy as np

# Charger ton fichier (remplace par ton vrai path)
df = pd.read_csv("../output/dataset_test_top40.csv")

df_synthetic = pd.DataFrame()

for col in df.columns:
    if df[col].dtype != "object":
        mean = df[col].mean()
        std = df[col].std()
        df_synthetic[col] = np.random.normal(mean, std, 2000)
    else:
        df_synthetic[col] = np.random.choice(df[col].dropna(), 2000)

# Sauvegarder
df_synthetic.to_csv("../output/dataset_test_top40_random.csv", index=False)

print("✅ 1000 lignes générées")
print(df_synthetic.shape)
```
